```python
def give_weights_to_classes(input_sequence):
    input_sequence = input_sequence.replace(' (?)', '')
    input_sequence = input_sequence.replace(' (väitöskirja?)', '')
    
    weights = {}
    
    if '\\' not in input_sequence and '/' not in input_sequence:
        weights[input_sequence] = 1.0
        
    elif '\\' in input_sequence and '/' in input_sequence:
        weights['vihkiminen'] = 1
        weights['virka'] = 1
        weights['viran vaihtaminen'] = 1
    
    elif '\\' in input_sequence:
        names = input_sequence.split('\\')

        for name in names:
            weights[name] = 1
    
    elif '/' in input_sequence:
        names = input_sequence.split('/')

        for name in names:
            weights[name] = 1
    
    return weights
```

```python
def weights_to_array(weights, feature_names):
    vector = np.zeros([len(feature_names),])
    for k, v in weights.items():
        vector[feature_names.index(k)] = v
    return vector
```

```python
multi_labels = []
for i in range(len(data_cut.Tapahtuma)):
    label = data_cut.Tapahtuma[i]
    weights = give_weights_to_classes(label)
    vec = weights_to_array(weights, list(all_names))
    multi_labels.append(vec)
```
